```python
from bs4 import BeautifulSoup 
import requests

url = 'https://shopping.interpark.com/best/main.do?&smid1=s_menu&smid2=best'
result = requests.get(url)
result_text = result.text
soup = BeautifulSoup(result_text, 'html.parser')

soup
```

```python
#상품 이름 파싱#
name = soup.select('a.name')
```

```python
for i in name:
    print(i.text.strip())
```

```python
#상품가격 파싱#
price = soup.select('span.number')
```

```python
for i in price:
    print(i.text.strip())
```

```python
# 100개만 출력해주는 경우

# 각 리스트 당 100개씩만 필요하므로 0번째부터 99까지 슬라이싱 => [:100]
# 이름과 가격을 나란히 출력시킴

for i in range(100):
    print(name[i].text, price[i].text)
```
