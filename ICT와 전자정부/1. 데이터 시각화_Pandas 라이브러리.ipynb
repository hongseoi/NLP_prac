###3) 막대그래프

```python
df.groupby('대업종')['광고집행금액'].mean().plot(kind='bar')
```

```python
df.groupby('대업종')['광고집행금액'].mean().plot(kind='barh')
```

```python
AVG = df.groupby('대업종')['광고집행금액'].mean()
```

```python
AVG = AVG.sort_values(ascending=False)
```

```python
AVG.plot(kind='barh')
```
