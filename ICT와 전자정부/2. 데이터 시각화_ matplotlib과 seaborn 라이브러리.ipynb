```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
tips
```

```python
sns.barplot(x='day', y='tip', data=tips)
```

```python
sns.barplot(x='day', y='tip', data=tips, palette='PuBu')
```

```python
sns.barplot(x='day', y='tip', data=tips, palette='PuBu')
```

```python
sns.barplot(x='time', y='tip', data=tips, palette='PuBu')
```

```python
sns.barplot(x='day', y='tip', data=tips, palette='Accent')  ## 위의 옵션에서 맞는 것을 선택함 
```

```python
sns.catplot(x='sex', y='total_bill', 
            data=tips, 
            kind='box')
```

```python
sns.catplot(x='sex', y='total_bill',
            hue='smoker', 
            data=tips, 
            kind='box')
```

```python
sns.catplot(x='sex', y='total_bill',
            hue='smoker', 
            col='time',
            data=tips, 
            kind='box')
```

- 점심에 비흡연자가 더 많은 비용을 지불한다
- 저녁에는 남성의경우 더 많이 지불하지만, 여성의 경우 큰 차이가 없다.

```python
sns.catplot(x='sex', y='tip',
            hue='smoker', 
            col='time',
            data=tips, 
            kind='box')
```

- 점심에 흡연자가 평균적으로 더 많은 팁을 지불한다
- 저녁에는 큰 차이가 없이 팁을 지불한다.
- 가장 많은 totalbill을 지불했던 dinner smoker male의 경우 지불한 tip 금액이 total bill과 패턴의 유사성이 없어 보인다.

###2) seaborn에서 제공하는 통계기반 시각화

```python
sns.violinplot(x='day', y='tip', data=tips, palette='PuBu')
```

```python
sns.violinplot(x='sex', y='total_bill', hue='smoker', data=tips, palette='PuBu')
```

```python
sns.lmplot(x='total_bill', y='tip', hue='sex', data=tips)
```
