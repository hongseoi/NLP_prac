
## 프로그래밍 기반 데이터 시각화와 실전 분석

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.barplot(x='day', y='tip', data=tips)
```

```python
sns.barplot(x='day', y='tip', data=tips, palette='PuBu')
```

```python
sns.barplot(x='day', y='tip', data=tips, palette='abc')
```

```python
sns.barplot(x='day', y='tip', data=tips, palette='Accent')  ## 위의 옵션에서 맞는 것을 선택함 
```

```python
sns.catplot(x='sex', y='total_bill', 
            data=tips, 
            kind='box')
```

```python
sns.catplot(x='sex', y='total_bill',
            hue='smoker', 
            data=tips, 
            kind='box')
```

```python
sns.catplot(x='sex', y='total_bill',
            hue='smoker', 
            col='time',
            data=tips, 
            kind='box')
```

```python
sns.catplot(x='sex', y='tip',
            hue='smoker', 
            col='time',
            data=tips, 
            kind='box')
```

###2) seaborn에서 제공하는 통계기반 시각화

```python
sns.violinplot(x='day', y='tip', data=tips, palette='PuBu')
```

```python
sns.violinplot(x='sex', y='total_bill', hue='smoker', data=tips, palette='PuBu')
```

```python
sns.lmplot(x='tip', y='total_bill', hue='sex', data=tips)
```

```python
sns.heatmap(tips.corr(), annot=True)
```

```python
sns.heatmap(tips.corr())
```
