1) 날짜 타입으로 변경하기

```python
import pandas as pd
```

```python
df = pd.read_csv('http://scottyoon.cafe24.com/data/brand_ad_ex.csv')
df
```

```python
df['시작일']
```

```python
df2 = df.copy()
df2['시작일'] = pd.to_datetime(df2['시작일'])  # dtype: object 를 datetime으로 변경함
df2['시작일']
```

```python
df2['시작일'].dt.dayofweek   #월요일: 0, 화: 1, 수:2, 목:3, 금: 4, 토: 5, 금: 6
```

```python
df2['시작_년'] = df2['시작일'].dt.year
df2['시작_월'] = df2['시작일'].dt.month
df2['시작_일'] = df2['시작일'].dt.day
```

```python
df2.head()
```

```python
df2.groupby('시작_월')['브랜드'].count()
```

```python
#df3 를 만들고 종료일 기준으로 날짜별로 브랜드의 카운트 시행
df3 = df.copy()
df3['종료일'] = pd.to_datetime(df3['종료일'])  # dtype: object 를 datetime으로 변경함df2['시작_년'] = df2['시작일'].dt.year
df3['종료_년'] = df3['종료일'].dt.year
df3['종료_월'] = df3['종료일'].dt.month
df3['종료_일'] = df3['종료일'].dt.day
df3.groupby('종료_일')['브랜드'].count()
```

```python
df3.groupby('종료_월')['Impression'].mean()
```

###2) 텍스트를 코드값으로 변경하기

```python
df3 = df.copy()
```

```python
df3['광고주국가'] = '대한민국'
```

```python
df3.loc[df3['광고주'] == '㈜ BBB', '광고주국가'] = '미국'
df3.loc[df3['광고주'] == '㈜ GGG', '광고주국가'] = '미국'
df3
```

```python
df3.loc[df3['광고주국가'] == '미국', '국가코드'] = 1
df3.loc[df3['광고주국가'] == '대한민국', '국가코드'] = 0
```

```python
df3.head()
```

```python
df3['국가코드'] = df3['국가코드'].astype(int)  # astype이면 정수로 바뀐다
df3.head()
```

```python
code = {
    '대한민국': 0,
    '미국': 1
}
```

```python
df3['국가코드2'] = df3['광고주국가'].map(code)
df3.head()
```

###3) 원핫인코딩

```python
df = pd.read_csv('http://scottyoon.cafe24.com/data/brand_ad_ex2.csv')
```

```python
df.head()
```

```python
df['업종'].value_counts()
```

```python
dummies = pd.get_dummies(df['업종'], drop_first=True) # 업종 중 마지막 종류인 건설업은 변수에서 제외하자!
```

```python
dummies
```

```python
df2 = df.copy()
```

```python
df2 = df2.join(dummies)
df2
```
