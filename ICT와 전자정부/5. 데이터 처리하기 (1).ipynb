### 2) 행열 추가 및 삭제

```
df3 = df.copy()
df3['CTR'] = round(df3['Click']/df3['Impression']*100, 2) # CTR 공식 적용 후 소수점 2자리로 변환 
```

```
df3.head()
```

```
df3['VTR'] = round(df3['재생 완료']/df3['Impression']*100, 3) # CTR 공식 적용 후 소수점 2자리로 변환
df3.head()
```

```
df3['광고주국가'] = '대한민국'
df3.head()
```

```
df3.loc[df3['광고주'] == '(주) BBB', '광고주국가'] = '미국'
df3
```

```
df3.loc[df3['Click'] > 5000, '광고주국가'] ='영국'
```

```
df3.loc[df3['광고주국가']=='영국'] # df3 중에서 광고주 국가가 영국인 경우만 출력해라 
```

```
df3
```

```
df3 = df3.drop([3, 11]) # 3번, 11번 행 삭제
df3
```

```
df3 =  df3[df3.광고주 != '(주)GGG'] # 광고주가 (주)GGG가 아닌 것만 추려라 
df3
```

```
df3.loc[(df3['광고주국가'] != '미국') & ~(df3['광고주국가'] != '영국')]
```

```
df3 = df3.drop('종료일', axis=1)
df3
```

###3) 그룹으로 묶어보기: Groupby

```
df3.groupby('광고주')['Impression'].mean() # 광고주별 데이터 평균
```

```
df3.groupby('시작일')['Click'].sum()
```
